# 재방문시 어느 지역에 자주 방문하는가?

```python
df_1 = df.copy() #Q1 범주(첫 방문/재방문)나눈 col추가한 df
bins_label = ['첫 방문', '재방문']
bins_range = [1, 2, df_1['Q1'].max()]
df_1["Q1_ob1"] = pd.cut(df_1["Q1"], bins_range, right=False, labels=bins_label)
```

```python
# Q1의 분포가 너무 넓어서 정규화
df_Q1_ob_temp1 = df_1[df_1['Q1']==1]
df_Q1_ob_temp2 = df_1[df_1['Q1']>1]

d_min = df_Q1_ob_temp2['Q1'].min(axis=0)
d_max = df_Q1_ob_temp2['Q1'].max(axis=0)
bins_label = [i for i in range(2,11)]

df_Q1_ob_temp1['Q1_ob2'] = 1

df_Q1_ob_temp2['Q1_ob_nor'] = (df_Q1_ob_temp2['Q1'] - d_min) / (d_max - d_min)
df_Q1_ob_temp2['Q1_ob2'] = pd.cut(df_Q1_ob_temp2['Q1_ob_nor'], 9, labels=bins_label)
df_Q1_ob_temp = pd.concat([df_Q1_ob_temp1, df_Q1_ob_temp2]).sort_index()

df_1 = pd.concat([df_1,df_Q1_ob_temp['Q1_ob2']], axis=1) #df_1에 정규화 범주 추가
```

```python
df_1['Q1'][df_1['Q1']>10].hist()
```

```python
# 사용할 columns : df['Q1_ob1', 'Q10_1B1~Q10_1B17']
temp = list(df_1.columns).index('Q10_1B1')
df1 = pd.concat([df_1.iloc[:,-2], df.iloc[:,temp:temp+17]], axis=1)
df1
```

```python
# 지역 방문한 횟수를 확인하기위해 count사용
df1_g = pd.DataFrame(df1.groupby('Q1_ob1').count())
#df1_g.plot.bar()
```

```python
df1_g.head()
```

```python
# 시각화용 df 생성
df1_gg = pd.DataFrame(np.zeros([17 * 2, 3]),
                      columns=['방문','지역','방문횟수'])
gg_lis = list(df1_g.index)
gg_dic = {1: '서울',2: '부산',3: '대구',
            4: '인천',5: '광주',6: '대전',
            7: '울산',8: '경기',9: '강원',
            10: '충북',11: '충남',12: '전북',
            13: '전남',14: '경북',15: '경남',
            16: '제주',17: '세종'}
gg_i = 0
for i in range(0, len(df1_g)):
    for j in range(0, len(df1_g.columns)):
        if i == 0:
            df1_gg.iloc[gg_i, 0] = gg_lis[i]
        else:
            df1_gg.iloc[gg_i, 0] = gg_lis[i]
            
        df1_gg.iloc[gg_i, 1] = j+1
        df1_gg.iloc[gg_i, 2] = df1_g.iloc[i, j]

        gg_i += 1
        
df1_gg = df1_gg.replace(gg_dic)
```

```python
# 시각화
plt.figure(figsize=(20,10))
plt.title('2017-2019년 방문객의 선호 방문지역', fontdict={'fontsize':30})
plt.xlabel('재방문 여부', fontsize=20)
plt.ylabel('', fontsize=20)

sns.barplot(x='방문',y='방문횟수',hue='지역',data=df1_gg)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=16,loc=1)
```

## 방문여부 범주화

```python
# 사용할 columns : df['Q1', 'Q9_2A1~Q9_2A3']
temp = list(df.columns).index('Q10_1B1')
df2 = pd.concat([df_1.iloc[:,[-1]], df_1.iloc[:,temp:temp+17]], axis=1)
df2
```

```python
# 지역 방문한 횟수를 확인하기위해 count사용
# 시각화용 df 생성
df2_g = pd.DataFrame(df2.groupby('Q1_ob2').count())

df2_gg = pd.DataFrame(np.zeros([len(df2_g) * len(df2_g.columns), 3]),
                      columns=['방문_범주','지역','방문횟수'])
gg_lis = list(df2_g.index)
gg_dic = {1: '서울',2: '부산',3: '대구',
            4: '인천',5: '광주',6: '대전',
            7: '울산',8: '경기',9: '강원',
            10: '충북',11: '충남',12: '전북',
            13: '전남',14: '경북',15: '경남',
            16: '제주',17: '세종'}
gg_i = 0
for i in range(0, len(df2_g)):
    for j in range(0, len(df2_g.columns)):
        df2_gg['방문_범주'][gg_i] = gg_lis[i]
        df2_gg['지역'][gg_i] = j+1
        df2_gg['방문횟수'][gg_i] = df2_g.iloc[i, j]

        gg_i += 1
        
df2_gg = df2_gg.replace({'지역':gg_dic})
```

```python
# 정규화로 인해 범주표시가 정확히 어려워 확인 후 대략적인 구간을 xlabel로 삽입

a = df_1[['Q1','Q1_ob2']]
a = pd.concat([a.groupby('Q1_ob2').min(),a.groupby('Q1_ob2').max()], axis=1)
a

#a를 기반으로 범주명 지정
x_label = ['첫 방문'] + ['{}이상\n{}이하'.format(i,j) for i, j in zip([2,36,70,102,150,200,210,251,300],[35,65,100,122,152,200,210,251,300])]
```

```python
# 시각화
plt.figure(figsize=(20,10))
plt.title('2017-2019년 방문객의 선호 방문지역', fontdict={'fontsize':30})

for i in range(0,len(x_label)):
    df2_gg = df2_gg.replace({'방문_범주':{i+1:x_label[i]}})

sns.barplot(x='방문_범주',y='방문횟수',hue='지역',data=df2_gg, palette='husl')

plt.xlabel('한국 방문횟수', fontsize=20)
plt.ylabel('지역 방문횟수', fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=20)
plt.ylim(0,4000) # 서울이 압도적으로 많아 y축 범위 설정
plt.legend(fontsize=16)
```

## 첫방문/재방문

```python
# 사용할 columns : df['Q1_ob1', 'CQ14_2_1~CQ14_2_6', CQ14_2_7', 'CQ14_2_10']
temp = list(df_2.columns).index('CQ14_2_1')
temp2 = list(df_2.columns).index('CQ14_2_7')
temp3 = list(df_2.columns).index('Q14_2_10')
df2 = pd.concat([df_2.iloc[:,-2], df_2.iloc[:,temp-1:temp+6],df_2.iloc[:,temp2], df_2.iloc[:,temp3]], axis=1)
df2.columns=['Q1_ob1', '전체 총액', '숙박비', '거주국 지불경비', '쇼핑비', '식음료비', '교통비', '여행사 지불비', '문화/오락비', '기타']
df2
```

```python
# 기타 변수가 csv저장 과정에서 파일이 깨져서 다시 (연속형 -> 범주형)과정 추가

df2_temp = df2[(df2['기타']>=1)&(df2['기타']<=4)|(df2['기타'] == 98) == False]

for i in df2_temp['기타'].index:
    if df2_temp['기타'][i] == 0:
        df2['기타'][i] = 1
    elif df2_temp['기타'][i] <=30:
        df2['기타'][i] = 2
    elif df2_temp['기타'][i] <=100:
        df2['기타'][i] = 3
    elif df2_temp['기타'][i]>10000:
        df2['기타'][i] = 98
    else:
        df2['기타'][i] = 4
display(df2, df2['기타'].value_counts())
```

- 시각화할 df 예상도


```python
df2_g = df2.copy()
df2_g = df2_g.astype({'숙박비':'object', '거주국 지불경비':'object', '쇼핑비':'object', '식음료비':'object', '교통비':'object', '여행사 지불비':'object', '문화/오락비':'object', '기타':'object'})
df2_g.info()
```

```python
# 시각화용 df 생성
df2_g=pd.DataFrame(np.zeros([80, 4]),
                      columns=['key','level','count', 'tot_mean'])

k=0
for val in df2.columns[2:]:
    df2_temp = df2[['Q1_ob1',val]].value_counts()
    df2_temp2 = df2[['전체 총액',val]].groupby([val]).mean()
    for i in ['첫 방문', '재방문']:
        for j in list(range(1,6))+[98]:
            df2_g['key'][k] = '({}, {})'.format(i,val)
            df2_g['level'][k] = j
            
            try:
                df2_g['count'][k] = df2_temp[i][j]
                df2_g['tot_mean'][k] = df2_temp2.loc[j,'전체 총액']
            except KeyError:
                pass
            k+=1
            
df2_g
```

```python
# 시각화
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(1,1,1)

plt.title('2017-2019년 재방문객의 소비금액', fontdict={'fontsize':30})

# 막대그래프
ax = sns.barplot(x='key',y='count',hue='level',data=df2_g)
plt.xlabel('')
plt.ylabel('', fontsize=20)
plt.xticks(rotation=75, fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=16, title = "범주별 소비금액", title_fontsize = 20)

# 선그래프
ax = ax.twinx()
a = df2_g[(df2_g['level']==5) | (df2_g['level']==98)]
ax = sns.lineplot(x='key',y='tot_mean', data=df2_g, ms=8, mfc='lightblue',
            color='pink', linestyle='--', lw=5, marker='o')

plt.ylim(0,895313)
plt.xlabel('',fontsize=18)
plt.ylabel('', fontsize=20)
plt.legend(['평균'], fontsize=16,bbox_to_anchor=(0.2,0.65),
           title = "전체 소비금액의 평균", title_fontsize = 20)
plt.yticks(fontsize=20)
```

- 첫방문일 때와 재방문일 때 소비패턴이 달라지진 않는다
- 거주국 지불경비, 교통비, 여행사 지불비, 문화/오락비, 기타에는 거의 금액을 소비하지 않는다
- 히트맵으로도 확인했을 때, 식음료비, 쇼핑비에 가장 많이 지출을 하는 것을 알 수 있다
- 초과금액은 쇼핑, 숙박비, 거주국 지불경비에만 있는 것을 알 수 있다

# 한국여행에 대한 항목별 만족도

```python
df_4 = df.copy()

# 사용할 columns : df['Q15_2A1~Q15_2A10']
temp = list(df_4.columns).index('Q15_2A1')
df4 = pd.concat([df_4.iloc[:, temp:temp + 10]], axis=1)

# 시각화용 df 생성
df4_g = None
a = ['출입국 절차', '대중교통', '숙박', '음식', '쇼핑', '관광지 매력도', '여행경비', '치안', '언어소통',
    '관광안내 서비스']
for val, b in zip(df4.columns, a):
    df_temp1 = pd.DataFrame(df4[val].value_counts())
    df_temp1['level'] = b
    df4_g = pd.concat([df4_g, df_temp1])
df4_g = df4_g.fillna(0)
df4_g = df4_g.reset_index()
```

```python
df4_dic = {1: '매우 불만족', 2: '대체로 불만족', 3: '보통', 4: '대체로 만족', 5: '매우 만족'}
df4_g = df4_g[df4_g['index'] > 0]
df4_g = df4_g.replace({'index': df4_dic})
```

```python
# 시각화
plt.figure(figsize=(20,10))
plt.title('한국여행에 대한 항목별 만족도', fontsize=30)

sns.barplot(x='level', y='count', hue='index', data=df4_g, palette='husl')

plt.xlabel('', fontsize=20)
plt.ylabel('')
plt.xticks(fontsize=17)
plt.yticks(fontsize=18)
plt.legend(fontsize=16)
```

# RFM _ 예측변수

```python
# 통합파일 불러오기
direct = '../Day_05/'
df = pd.read_csv(direct+'2017-2019_통합본.csv')
df = df.drop(df[df['Q1'] > 300].index).reset_index(drop = True)

df_1 = df[['D_GEN', 'D_AGE', 'D_COU1', 'Q1', 'Q14_1T', 'Q16', 'ID', 'D_YEAR']]
df_1['D_YEAR'] = df_1['D_YEAR'].str[:4]
df_1 = df_1.reset_index(drop=True)
```

```python
# 재방문 예측 데이터 만들기
df_rfm = pd.DataFrame()

# 예측 라벨 만들기
p_list = ['전혀 그렇지 않다', '그렇지 않다', '보통', '그렇다', '매우 그렇다']
r_labels = range(1, 6)

# 예측 데이터 만들기
cut = pd.cut(df_1['Q1'], bins = [0, 1, 3, 5, 10, 300])
df_rfm['최근방문'] = df_1['D_YEAR']
df_rfm['최근방문'] = df_rfm['최근방문'].replace({'2017' : 1, '2018' : 3, '2019' : 5})
df_rfm['방문횟수'] = pd.cut(df_1['Q1'], bins = [0, 1, 3, 5, 10, 300], labels = r_labels)
df_rfm['구매금액'] = pd.cut(df_1['Q14_1T'], bins = [0, 2500, 5000, 7500, 10000, df_1['Q14_1T'].max()], labels = r_labels)

df_rfm['최근방문'] = df_rfm['최근방문'].astype(str)
df_rfm['방문횟수'] = df_rfm['방문횟수'].astype(str)
df_rfm['구매금액'] = df_rfm['구매금액'].astype(str)
df_rfm['점수'] = df_rfm['최근방문'] + df_rfm['방문횟수'] + df_rfm['구매금액']
df_rfm['구매금액'] = df_rfm['구매금액'].astype(int)
df_rfm.head()
```

## rfm_R, rfm_F, rfm_M 각각 재방문 예측정도 확인

```python
df_rfm['최근방문'] = df_rfm['최근방문'].astype(int) #R
df_rfm['방문횟수'] = df_rfm['방문횟수'].astype(int) #F
df_rfm['구매금액'] = df_rfm['구매금액'].astype(int) #M

df_rfm_2 = df_rfm.copy()
df_rfm_2[:5]
```

```python
df_rfm_2['범주'] = '재방문예측'

# R
df_rfm_23=df_rfm_2.copy()
df_rfm_23['최근방문'] = df_1['Q16']
df_rfm_23['범주'] = '재방문의향'

df_rfm_3_g = pd.concat([df_rfm_2, df_rfm_23]).reset_index(drop = True)
df_rfm_3_g.head()
```

```python
plt.rcParams['figure.figsize'] = (20, 10)
plt.title('2017-2019년 외래방문객 재방문예측/의향 비교', fontdict={'fontsize': 30})
ax = sns.countplot(x='최근방문', hue='범주', palette='husl', data=df_rfm_3_g)
ax.set_xticklabels(['전혀 그렇지 않다', '그렇지 않다', '보통', '그렇다', '매우 그렇다'],
                   fontdict={'fontsize': 20})
plt.yticks(fontsize=20)
plt.xlabel('5점척도', fontdict={'fontsize': 20})
plt.ylabel('', fontdict={'fontsize': 20})
plt.legend(fontsize=16,loc=1)
```

```python
# F
df_rfm_24=df_rfm_2.copy()
df_rfm_24['방문횟수'] = df_1['Q16']
df_rfm_24['범주'] = '재방문의향'

df_rfm_4_g = pd.concat([df_rfm_2, df_rfm_24]).reset_index(drop = True)
```

```python
plt.rcParams['figure.figsize'] = (20, 10)
plt.title('2017-2019년 외래방문객 재방문예측/의향 비교', fontdict={'fontsize': 30})
ax = sns.countplot(x='방문횟수', hue='범주', palette='husl', data=df_rfm_4_g)
ax.set_xticklabels(['전혀 그렇지 않다', '그렇지 않다', '보통', '그렇다', '매우 그렇다'],
                   fontdict={'fontsize': 20})
plt.yticks(fontsize=20)
plt.xlabel('5점척도', fontdict={'fontsize': 20})
plt.ylabel('', fontdict={'fontsize': 20})
plt.legend(fontsize=16,loc=1)
```

```python
# M
df_rfm_25=df_rfm_2.copy()
df_rfm_25['구매금액'] = df_1['Q16']
df_rfm_25['범주'] = '재방문의향'

df_rfm_5_g = pd.concat([df_rfm_2, df_rfm_25]).reset_index(drop = True)
```

## RFM 통합해서 시각화

```python
# 예측
a = df_rfm_2[['최근방문']]
a = a.rename(columns={'최근방문':'점수'})
a['level'] = 'R'

b = df_rfm_2[['방문횟수']]
b = b.rename(columns={'방문횟수':'점수'})
b['level'] = 'F'

c = df_rfm_2[['구매금액']]
c = c.rename(columns={'구매금액':'점수'})
c['level'] = 'M'

df_rfm_6 = pd.concat([a,b,c])
```

```python
plt.rcParams['figure.figsize'] = (20, 10)

plt.title('RFM점수별 재방문예측', fontdict={'fontsize': 30})

ax = sns.countplot(x='level', hue='점수', palette='husl', data=df_rfm_6)

ax.set_xticklabels(['R', 'F', 'M'], fontdict={'fontsize': 20})
plt.yticks(fontsize=20)
plt.xlabel('')
plt.ylabel('')
plt.legend(('전혀 그렇지 않다', '그렇지 않다', '보통', '그렇다', '매우 그렇다'), fontsize=16,loc=1)
```

# RFM 예측변수

## weight = 5

```python
# 가중치 (최근방문 = 5>방문횟수 = 2>구매금액 = 1)
df_rfm_w = df_rfm.copy()
df_rfm['최근방문'] = df_rfm['최근방문'].astype(int)
df_rfm['방문횟수'] = df_rfm['방문횟수'].astype(int)
df_rfm['구매금액'] = df_rfm['구매금액'].astype(int)

df_rfm_w['최근방문'] = df_rfm['최근방문']*5
df_rfm_w['방문횟수'] = df_rfm['방문횟수']*2
df_rfm_w['구매금액'] = df_rfm['구매금액']
df_rfm_w['점수'] = df_rfm_w['최근방문']+df_rfm_w['방문횟수']+df_rfm_w['구매금액']

# 동일한 길이로 5등분 (5점척도)
df_rfm_w['재방문예측'] = pd.cut(df_rfm_w['점수'], bins = 5, labels = r_labels)
df_rfm_w['범주'] = '재방문예측'

df_rfm_w2=df_rfm_w.copy()
df_rfm_w2['재방문예측'] = df_1['Q16']
df_rfm_w2['범주'] = '재방문의향'

# 시각화용 df 생성
df_rfm_w_g = pd.concat([df_rfm_w, df_rfm_w2]).reset_index(drop = True)
```

```python
df_rfm_w_g[:5]
```

```python
# 시각화
plt.rcParams['figure.figsize'] = (20, 10)
plt.title('2017-2019년 외래방문객 재방문예측/의향 비교', fontdict={'fontsize': 30})
ax = sns.countplot(x='재방문예측', hue='범주', palette='husl', data=df_rfm_w_g)

ax.set_xticklabels(['전혀 그렇지 않다', '그렇지 않다', '보통', '그렇다', '매우 그렇다'],
                   fontdict={'fontsize': 20})
plt.yticks(fontsize=20)
plt.xlabel('5점척도', fontdict={'fontsize': 20})
plt.ylabel('', fontdict={'fontsize': 20})
plt.legend(fontsize=16,loc=1)

a = len(df_rfm_w_g[(df_rfm_w_g['재방문예측'] == 1) & (df_rfm_w_g['범주'] == '재방문예측')])
plt.axhline(a, linestyle='--', color='red')
```

## 재방문 할것으로 예측되는 고객의 인적사항

```python
df_rfm_w[:5]
```

```python
df_2 = df_1[['D_GEN', 'D_AGE', 'D_COU1']]

df_2_m = pd.concat([df_rfm_w, df_2], axis=1)
df_2_m = df_2_m[df_2_m['재방문예측']>=4]
```

```python
dic1 = {1: '남성', 2: '여성'}

dic2 = {1: '15-20세',
    2: '21-30세',
    3: '31-40세',
    4: '41-50세',
    5: '51-60세',
    6: '61세 이상'}

dic3 = {1: '일본',
    2: '중국',
    3: '홍콩',
    4: '싱가포르',
    5: '대만',
    6: '태국',
    7: '말레이시아',
    8: '호주',
    9: '미국',
    10: '캐나다',
    11: '영국',
    12: '독일',
    13: '프랑스',
    14: '러시아',
    15: '중동',
    16: '인도',
    17: '필리핀',
    18: '인도네시아',
    19: '베트남',
    97: '기타'}

df_2_m = df_2_m.replace({'D_GEN':dic1, 'D_AGE':dic2, 'D_COU1':dic3})
```

```python
# 시각화용 df 생성
df_2_m_g = df_2_m.groupby(['D_GEN', 'D_AGE', 'D_COU1'])
df_2_m_count = pd.DataFrame(df_2_m_g['점수'].count())

a = pd.DataFrame(df_2_m_count.values)

df_2_m_temp = pd.DataFrame(df_2_m_count.index)
df_2_m_temp = pd.concat([a, df_2_m_temp],axis=1)
df_2_m_temp.columns = ['count', 'cate']
df_2_m_temp = df_2_m_temp.sort_values('count', ascending=False)[:20]
df_2_m_temp
```

```python
# 시각화
plt.rcParams['figure.figsize'] = (20, 10)
plt.title('재방문 할 것으로 예측되는 방문객', fontdict={'fontsize': 30})
sns.barplot(data = df_2_m_temp,x="cate",y="count")
plt.xticks(rotation=75, fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('')
plt.ylabel('')
plt.legend(fontsize=16,loc=1)
```

```python
# Top20 중 나이 범수개수 카운트
df_2_m_g = df_2_m.groupby(['D_GEN', 'D_COU1', 'D_AGE'])
df_2_m_count = pd.DataFrame(df_2_m_g['점수'].count())
b = df_2_m_count.sort_values('점수', ascending=False)[:20]
b = b.unstack()
b.count()
```

```python
# Top20 중 성별 범수개수 카운트
df_2_m_g = df_2_m.groupby(['D_AGE', 'D_COU1', 'D_GEN'])
df_2_m_count = pd.DataFrame(df_2_m_g['점수'].count())
b = df_2_m_count.sort_values('점수', ascending=False)[:20]
b = b.unstack()
b.count()
```
